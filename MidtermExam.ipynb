```python
############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
#  ADD HERE
###########################################################

############################################################
#  ListNode class
#  NOTHING CAN BE CHANGED IN THIS CLASS
###########################################################
class ListNode:
    def __init__(self, val = 0, next= None):
        self.val = val
        self.next = next

    def change_value(self,x:'int')->'None':
        self.val = x

############################################################
#  Slist class
#  WRITE CODE ONLY IN SLIST CLASS
###########################################################
class Slist:
    def __init__(self, n:'Python Int' = None):
        #You must use first and last.
        self._first = None
        self._last = None
        
        ## add what ever you want here to handle sign   
        # YOU CANNOT STORE n in SLIST DATA STRUCTURE   
        #construct Slist using n, digit by digit
        #if n is None, both _first and _last is None
        # if n = 1986 1->9->8->6->None
        # You cannot convert slist to PYTHON INT or any other data structure

        self._is_positive = True  # Default to positive

        if (n is not None):
            self.build(n)

    
    ########################################################### 
    # TIME:O(log to base 10 of n)
    # SPACE:O(log to base 10 of n)
    ###########################################################         
    def build(self, n: 'Python int') -> None:
        if (n < 0):
            self._is_positive = False
            n = -n  # Make n positive for processing

        if (n == 0):
            self.prepend(0)
        else:
            while (n != 0):
                self.prepend(n % 10)
                n = n // 10

    ########################################################### 
    # Prepend: add a node at the begining of slist
    # Time: THETA(1)
    # Space: THETA(1)
    ########################################################### 
    def prepend(self, val: 'int') -> 'None':
        new_node = ListNode(val)
        if (self._first is None):
            self._first = new_node
            self._last = new_node
        else:
            new_node.next = self._first
            self._first = new_node
  

    ######################################################
    # This routine is only for testing. You cannot use it
    # You cannot convert slist to PYTHON INT or any other data structure
    ####################################################
    def Int(self)->'python int':
        s = 0
        sign = self.is_positive()
        n = self._first 
        assert(n)
        l = len(self)
        if (l > 1):
          if (n.val == 0):
            print(self)
            print("You cannot have leading zero")
            assert(0)
        if (l == 1 and n.val == 0):
            if (sign == False):
                print(self)
                print("There is no -0")
                assert(0)
        
        while (n != None):
          v = n.val 
          if (v < 0 or v > 9):
            print(self)
            print("YOU MUST STORE SINGLE DIGIT 0 to 9 only with each node")
            assert(0)
          s = (10 * s)  + v
          n = n.next
        if (sign == False):
          s = -s
        return s
       
        
    #############################
    # WRITE All public functions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    #############################

    #############################
    # YOU MUST WRITE whether the number represented in Slist + or -
    # NOTE THERE IS NO -0. For 0 sign must be positive
    #############################
    def is_positive(self)->'Bool':
        return self._is_positive
        

    def is_negative(self)->'Bool':
        return not self.is_positive()
    ########################################################### 
    # Get number of elements in the slist
    # Time: THETA(n)
    # Space: THETA(1)
    ###########################################################    
    def __len__(self)->'int':
        l = 0
        curr = self._first
        while (curr is not None):
            l = l + 1
            curr = curr.next
        return l
        

    ########################################################### 
    # Time: O(n)
    # Space: O(n)
    ###########################################################  
    def __str__(self) -> 'str':
        s = ""
        if (not self.is_positive()):
            s += "-"
        else: 
            s += "+"
        curr = self._first
        while (curr):
            s += str(curr.val)
            if (curr.next):
                s += "->"
            curr = curr.next
        s += "->None"
        return s


########################################
# Calling code
########################################
class Exam:
    def __init__(self, a:'Slist', b:'Slist', clist:'Empty Slist', work:'list of size 1 that has int', show:'Bool')->'None':
        self.a = a
        self.b = b
        self.clist = clist
        self.work = work
        self.show = show

        if (show):
          print("a =",a)
          print("b =",b)

        self._add(a, b, clist, work)

    ########################################################### 
    # Time: THETA(n)
    # Space: THETA(1)
    ########################################################### 
    def _add(self, a: 'Slist', b: 'Slist', clist: 'Empty Slist', work: 'list of size 1 that has int') -> 'None':
        if (a.is_positive() == b.is_positive()):
            self._addition(a, b, clist, work)
            clist._is_positive = a.is_positive()
        else:
            if (self._compare(a, b) >= 0):
                self._subtraction(a, b, clist, work)
                clist._is_positive = a.is_positive()
            else:
                self._subtraction(b, a, clist, work)
                clist._is_positive = b.is_positive()
        
        # Ensure that zero is always positive
        if ((clist.__len__() == 1) and (clist._first.val == 0)):
            clist._is_positive = True

        # Remove leading zeros
        self._remove_leading_zeros(clist)
                
    ########################################################### 
    # Time: THETA(n)
    # Space: THETA(1)
    ########################################################### 
    def _addition(self, a: 'Slist', b: 'Slist', clist: 'Empty Slist', work: 'list of size 1 that has int') -> 'None':
        carry = 0
        a1 = self._reverse_list(a._first)
        b1 = self._reverse_list(b._first)

        while ((a1 is not None) or (b1 is not None) or (carry != 0)):
            val1 = a1.val if a1 is not None else 0
            val2 = b1.val if b1 is not None else 0

            total = val1 + val2 + carry
            carry = total // 10
            self._append(clist, total % 10)

            if (a1 is not None):
                a1 = a1.next
            if (b1 is not None):
                b1 = b1.next

            work[0] += 1

        clist._first = self._reverse_list(clist._first)
        
    ########################################################### 
    # Time: THETA(n)
    # Space: THETA(n)
    ###########################################################  
    def _subtraction(self, a: 'Slist', b: 'Slist', clist: 'Empty Slist', work: 'list of size 1 that has int') -> 'None':
        borrow = 0
        a1 = self._reverse_list(a._first)
        b1 = self._reverse_list(b._first)

        while ((a1 is not None) or (b1 is not None) or (borrow != 0)):
            val1 = a1.val if a1 is not None else 0
            val2 = b1.val if b1 is not None else 0

            total = val1 - val2 - borrow
            if (total < 0):
                total += 10
                borrow = 1
            else:
                borrow = 0
            self._append(clist, total)

            if (a1 is not None):
                a1 = a1.next
            if (b1 is not None):
                b1 = b1.next

            work[0] += 1

        clist._first = self._reverse_list(clist._first)
        
    ########################################################### 
    # Time: THETA(n)
    # Space: THETA(1)
    ###########################################################  
    def _compare(self, a: 'Slist', b: 'Slist') -> 'int':
        len_a = len(a)
        len_b = len(b)
        if len_a != len_b:
            return len_a - len_b
        a1 = a._first
        b1 = b._first
        while ((a1 is not None) and (b1 is not None)):
            if (a1.val != b1.val):
                return a1.val - b1.val
            a1 = a1.next
            b1 = b1.next
        return 0
        
    ########################################################### 
    # Time: O(n)
    # Space: O(1)
    ###########################################################  
    def _reverse_list(self, head: 'ListNode') -> 'ListNode':
        prev = None
        curr = head
        while (curr is not None):
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        return prev


    ########################################################### 
    # Append: add a node at the end of slist
    # Time: THETA(1)
    # Space: THETA(1)
    ########################################################### 
    def _append(self, clist: 'Slist', val: 'int') -> 'None':
        new_node = ListNode(val)
        if (clist._first is None):
            clist._first = new_node
            clist._last = new_node
        else:
            clist._last.next = new_node
            clist._last = new_node
            
    ########################################################### 
    # Time: O(n)
    # Space: O(1)
    ###########################################################  
    def _remove_leading_zeros(self, clist: 'Slist') -> 'None':
        # Remove leading zeros in the resulting list
        while ((clist._first is not None) and (clist._first.val == 0) and (clist.__len__() > 1)):
            clist._first = clist._first.next
        if (clist._first is None):
            clist._last = None




```
